```python
# As we have the cnt, we can load from file
def load_cnt_fare(file, start, end, cnt_file, total_fare_file):
  df = pd.read_csv(file)
  cnt = np.zeros((7*24, 263))
  total_fare = np.zeros((7*24, 263))
  
  for index, row in df.iterrows():
    if index % 100000 == 0:
      print(index, df.shape[0])
    
    pickup_datetime = datetime.datetime.strptime(row['tpep_pickup_datetime'], '%Y-%m-%d %H:%M:%S')
    dropoff_datetime = datetime.datetime.strptime(row['tpep_dropoff_datetime'], '%Y-%m-%d %H:%M:%S')
    
    if row['tpep_pickup_datetime'] >= start and row['tpep_pickup_datetime'] < end and int(row['PULocationID']) <= 263 and int(row['DOLocationID']) <= 263:
#       if (pickup_datetime - datetime.datetime.strptime('2019-01-01', '%Y-%m-%d')).days % 2 == 0:
      row_num = pickup_datetime.weekday() * 24 + pickup_datetime.hour
      col_num = int(row['PULocationID']) - 1
      cnt[row_num][col_num] += 1
      total_fare[row_num][col_num] += float(row['total_amount'])
  np.save(cnt_file, cnt)
  np.save(total_fare_file, total_fare)  
```

```python
def new_algo_test(file, start, end, save_file_name):
  global prev_new, test_profit_new, max_profit_id
  df = pd.read_csv(file)
  print("data loaded")
  df = df[(df.tpep_pickup_datetime >= start) & (df.tpep_pickup_datetime < end)]
  df = df.sort_values(by=['tpep_pickup_datetime'])
  
  final_time = datetime.datetime.strptime(end, '%Y-%m-%d %H:%M:%S')
  with open(save_file_name, "a+") as f:
    while prev_new[0] < end:
      ele = prev_new
      f.write(str(ele))
      f.write(" ")
      f.write(str(test_profit_new))
      f.write("\n")
      t = datetime.datetime.strptime(ele[0], '%Y-%m-%d %H:%M:%S')
      weekday = t.weekday()
      day = t.day
      hour = t.hour
      row_idx = weekday * 24 + hour

      max_ids = set(np.argsort(-fare[row_idx])[:1] + 1)

      df_filter = df[(df.tpep_pickup_datetime > ele[0]) & (df.PULocationID.isin(max_ids))].reset_index()
      found = True
      j = 0
      if df_filter.shape[0] == 0:
  #       print("Category 1")
        found = False
      else:
  #       print("Category 2")
        new_t = datetime.datetime.strptime(df_filter.iloc[0]['tpep_pickup_datetime'], '%Y-%m-%d %H:%M:%S')
        new_weekday = new_t.weekday()
        new_day = new_t.day
        new_hour = new_t.hour
      while ((not found) or (new_hour != hour or new_day != day)) and (new_t < final_time):
  #       if not found:
  #         print("not found")
  #       elif new_day != day:
  #         print("day not match", "new_day", new_day, "day", day)
  #       elif new_hour != hour:
  #         print("hour not match", "hour", hour, "new_hour", new_hour)
  #       else:
  #         print("???")
        if day > 31:
          return
        j = 0
        hour += 1
        if hour == 24:
          hour = 0
          weekday += 1
          weekday %= 7
          day += 1
        row_idx = weekday * 24 + hour

        max_ids = set(np.argsort(-fare[row_idx])[:1] + 1)

        df_filter = df[(df.tpep_pickup_datetime > ele[0]) & (df.PULocationID.isin(max_ids))].reset_index()
        if df_filter.shape[0] == 0:
  #         print("Category 4")
          found = False
        else:
  #         print("Category 4")
          found = True
          for i in range(df_filter.shape[0]):
            new_t = datetime.datetime.strptime(df_filter.iloc[i]['tpep_pickup_datetime'], '%Y-%m-%d %H:%M:%S')
            new_weekday = new_t.weekday()
            new_day = new_t.day
            new_hour = new_t.hour
            if new_day < day or (new_day == day and new_hour < hour):
  #             print("Loop here")
              continue
            else:
              j = i
              break

      if (new_t >= final_time):
        break
      new_ele = (df_filter.iloc[j]['tpep_dropoff_datetime'], df_filter.iloc[j]['DOLocationID'])
      test_profit_new += float(df_filter.iloc[j]['total_amount'])
      prev_new = new_ele
```
